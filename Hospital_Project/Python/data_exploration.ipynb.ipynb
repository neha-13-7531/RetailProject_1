```python
# cloud setup and data lake
import os
import shutil
import pandas as pd
```

```python
# Create Data Lake Folder Structure
base_folder = "MediFlow_DataLake"

raw_folder = os.path.join(base_folder, "Raw")
staging_folder = os.path.join(base_folder, "Staging")
processed_folder = os.path.join(base_folder, "Processed")

os.makedirs(raw_folder, exist_ok=True)
os.makedirs(staging_folder, exist_ok=True)
os.makedirs(processed_folder, exist_ok=True)

print("Data Lake folders created successfully")
```

```python
# Upload Dataset to RAW Zone
source_file = "C:\\Users\\Abc\\Downloads\\diabetic_data.csv"   
destination_file = os.path.join(raw_folder,"diabetic_data.csv")

shutil.copy(source_file, destination_file)

print("Your Dataset uploaded to RAW zone")
```
