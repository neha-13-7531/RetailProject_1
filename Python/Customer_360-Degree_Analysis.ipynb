```python
df = df.dropna(subset=['Customer ID']) # remove the null values in customer id column
df
```

```python
df = df[df['Price'] > 0] # get price greter than 0 
df
```

```python
df = df[df['Quantity'] > 0]
df
```

```python
df.loc[:, 'InvoiceDate'] = pd.to_datetime(
    df['InvoiceDate'],
    format="%d-%m-%Y %H:%M"
)
```

```python
# Remove negative quantity
df = df[df['Quantity'] > 0].copy()

# Convert datatypes
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceDate']
```

```python
df = df[df['Quantity'] > 0].copy()
```

```python
# Convert CustomerID to string
df['Customer ID'] = df['Customer ID'].astype(str)
df['Customer ID']
```

```python
df.dtypes # check datype of each column
```

```python
# Create Total Amount column
df['TotalAmount'] = df['Quantity'] * df['Price']

df.head()
```

```python
total_sales = df['TotalAmount'].sum() # Total sale
total_sales
```

```python
total_customers = df['Customer ID'].nunique()  # total unique customer
total_customers
```

```python
country_sales = df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False)  # get top 10 country by sale

country_sales.head(10)
```

```python
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False) # Top 10 Products by Quantity Sold

top_products.head(10)
```

```python
# monthly sale trend
# Extract Month
df['Month'] = df['InvoiceDate'].dt.to_period('M')
df['Month']
```

```python
monthly_sales = df.groupby('Month')['TotalAmount'].sum()
monthly_sales
```

```python
# Plot
plt.figure(figsize=(10,5))
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```
